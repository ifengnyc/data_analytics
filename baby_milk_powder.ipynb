```
import pandas as pd
```

```
data_url = 'https://data.fda.gov.tw/opendata/exportDataList.do?method=ExportData&InfoId=11&logType=2'
df = pd.read_csv(data_url)
df.sample(1)
```

```
df['申請商名稱'].value_counts()
```

```
df.groupby(['申請商名稱','產地'])['中文名稱'].count().to_frame()
```

```
df['成分'].str.split(';').explode().value_counts()
```

|特點|魚油|藻油|
|----|----|----|
|來源|富含脂肪的魚類|海洋微藻|
|適合對象|大多數嬰兒，但避免魚類過敏|所有嬰兒，包括魚類過敏和素食者|
|DHA 含量|較高|較低，但足夠嬰兒需求|
|EPA 含量|可能含有|不含|
|環境影響|可能涉及過度捕撈|更可持續|
